```python

quality_columns = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 'Clean.Cup', 'Sweetness', 'Cupper.Points']
a['Total.Quality.Score'] = a[quality_columns].sum(axis=1)

a[['Total.Quality.Score']].head()
```

```python

a['Mean.Altitude'] = a[['altitude_low_meters', 'altitude_high_meters', 'altitude_mean_meters']].mean(axis=1)

a[['Mean.Altitude']].head()
```

```python

a['Defect.Ratio'] = a['Category.One.Defects'] / (a['Category.Two.Defects'] + 1)


a[['Defect.Ratio']].head()
```

```python

quality_threshold = 50
filtered_data_quality = a[a['Total.Quality.Score'] > quality_threshold]

filtered_data_quality.head()
```

```python

a['Grading.Date'] = pd.to_datetime(a['Grading.Date'], errors='coerce')

start_date = '2015-01-01'
end_date = '2020-12-31'
filtered_data_date = a[(a['Grading.Date'] >= start_date) & (a['Grading.Date'] <= end_date)]

filtered_data_date.head()
```
